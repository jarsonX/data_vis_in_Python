## Line Charts
### Data Visualization with Python example

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Load files, change data types
df_sa = pd.read_fwf('52_sa_median_solution.txt', infer_nrows=100, names=['sa'])
sa = df_sa['sa'].str.replace(',','.').astype(np.float64)

df_secrest = pd.read_fwf('52_secrest_median_solution.txt', infer_nrows=100, names=['secrest'])
secrest = df_secrest['secrest'].str.replace(',','.').astype(np.float64)
```

```python
#Plot data
plt.style.use('seaborn')

figure, axes = plt.subplots()

axes.plot(sa, label='SA', color='gray', linewidth=1)
axes.plot(secrest, label='Secrest', color='green', linewidth=3.0)
axes.legend(loc='upper right')

plt.title('Algorithms Comparison', fontsize=16)
plt.ylabel('Solution (distance)')
plt.xlabel('Number of iterations')

plt.savefig('case_2_line_chart.png', dpi=(300))
plt.show()
```
