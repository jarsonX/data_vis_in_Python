## 3D Charts
### Data Visualization with Python example

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Define functions
def alpine_no_2(x,y):
    return np.cos(x) * x * (1 + (np.cos(y)/2))

def schubert(x,y):
    return ((np.cos(2*x+1)) + (2*np.cos(3*x+2)) 
           + (3*np.cos(4*x+3)) + (4*np.cos(5*x+4))
           + (5*np.cos(6*x+5))) * ((np.cos(2*y+1))
           + (2*np.cos(3*y+2)) + (3*np.cos(4*y+3))
           + (4*np.cos(5*y+4)) + (5*np.cos(6*y+5)))

def rastring(x,y):
    return (x*x - np.cos(18*np.pi*x)) + (y*y - np.cos(18*np.pi*y))

#Function for plotting
def plot3D(X,Y,Z,filename,style='viridis'):
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.plot_surface(X,Y,Z, cmap=style)

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

    plt.tight_layout()
    plt.savefig(filename, dpi=(300))
    plt.show()
```

```python
#Generating input values
x = np.linspace(0, 12, 30)
y = np.linspace(0, 12, 30)
```

```python
#Plot 1: Alpine No 2
X, Y = np.meshgrid(x, y)
Z = alpine_no_2(X, Y)

plot3D(X,Y,Z,'case_12_3D_alpine_no_2.jpg')
```

```python
#Plot 2: Schubert
X, Y = np.meshgrid(x, y)
Z = schubert(X, Y)

plot3D(X,Y,Z,'case_12_3D_schubert.jpg')
```

```python
#Plot 3: Rastring
X, Y = np.meshgrid(x, y)
Z = rastring(X, Y)

plot3D(X,Y,Z,'case_12_3D_rastring.jpg')
```
