## Scatter Charts
### Data Visualization with Python example

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load data
assets = pd.read_table('assets.txt')
pareto = pd.read_table('Pareto Front.txt')

pareto_high = pareto.head(500)
pareto_low = pareto.tail(500)
pareto_rest = pareto[500:-500]
```

```python
#Plot data
plt.style.use('seaborn')

figure, axes = plt.subplots()
figure.set_size_inches(12,8)

axes.scatter(assets.Variance, assets.Mean, color='gray', label='Assets')
axes.scatter(pareto.Variance, pareto.Mean, color='lightgray')
axes.scatter(pareto_high.Variance, pareto_high.Mean, color='red', label='High risk & return')
axes.scatter(pareto_low.Variance, pareto_low.Mean, color='green', label='Low risk & return')
axes.scatter(pareto.median()[0], pareto.median()[1], color='black', sizes=(200,200), label='Median')

axes.legend(loc='upper right')

plt.title('Pareto Front', fontsize=16)
plt.ylabel('Mean')
plt.xlabel('Variance')
plt.text(0.0001, -0.0005, 'LOW', weight='semibold', size='medium')
plt.text(0.0015, 0.0042, 'HIGH', weight='semibold', size='medium')
plt.text(0.000510, 0.0019, 'MEDIAN', weight='semibold', size='medium') 

plt.savefig('case_3_scatter_chart.png', dpi=(300))
plt.show()
```
