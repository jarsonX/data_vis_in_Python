## Line Charts
### Data Visualization with Python example

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load data
df = pd.read_csv('input_data_7.csv', sep=';', decimal=',')
```

```python
#Function to plot charts
def plot_chart(row, column, title, axes, data):
    axes[row][column].set_title(title, fontsize=12)
    axes[row][column].set_xlabel('Number of iterations', fontsize=10)
    axes[row][column].set_ylabel('Solution', fontsize=10)
    axes[row][column].set(ylim=(0,4))  # the same scale for comparison
    axes[row][column].plot(data)
```

```python
plt.style.use('seaborn')

figure, axes = plt.subplots(4,2)
figure.set_size_inches(12,10)

plot_chart(0, 0, 'Algorithm 1', axes, df['Algorytm 1'])
plot_chart(1, 0, 'Algorithm 2', axes, df['Algorytm 2'])
plot_chart(2, 0, 'Algorithm 3', axes, df['Algorytm 3'])
plot_chart(3, 0, 'Algorithm 4', axes, df['Algorytm 4'])
plot_chart(0, 1, 'Algorithm 5', axes, df['Algorytm 5'])
plot_chart(1, 1, 'Algorithm 6', axes, df['Algorytm 6'])
plot_chart(2, 1, 'Algorithm 7', axes, df['Algorytm 7'])
plot_chart(3, 1, 'Algorithm 8', axes, df['Algorytm 8'])

plt.tight_layout()
plt.savefig('case_7_line_chart.png', dpi=(300))
plt.show()
```
