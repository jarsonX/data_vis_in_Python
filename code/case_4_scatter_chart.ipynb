## Scatter Charts
### Data Visualization with Python example

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load data
attributes = ['sep_len', 'sep_wid', 'pet_len', 'pet_wid']
iris = pd.read_csv('iris.data', header=None, names=attributes, usecols=[0,1,2,3])

sep_len = iris.sep_len
sep_wid = iris.sep_wid
pet_len = iris.pet_len
pet_wid = iris.pet_wid
```

```python
#Plot data
plt.style.use('seaborn')

figure, axes = plt.subplots(2,2)
figure.set_size_inches(12,8)

axes[0][0].scatter(sep_len, sep_wid, color='mediumblue')
axes[0][0].set_title('Sepal Length - Sepal Width')
axes[0][0].set_xlabel('Sepal Length')
axes[0][0].set_ylabel('Sepal Width')

axes[0][1].scatter(pet_len, pet_wid, color='mediumslateblue')
axes[0][1].set_title('Petal Length - Petal Width')
axes[0][1].set_xlabel('Petal Length')
axes[0][1].set_ylabel('Petal Width')

axes[1][0].scatter(sep_len, pet_wid, color='deeppink')
axes[1][0].set_title('Sepal Length - Petal Width')
axes[1][0].set_xlabel('Sepal Length')
axes[1][0].set_ylabel('Petal Width')

axes[1][1].scatter(pet_len, sep_wid, color='orangered')
axes[1][1].set_title('Petal Length - Sepal Width')
axes[1][1].set_xlabel('Petal Length')
axes[1][1].set_ylabel('Sepal Width')

plt.tight_layout()
plt.savefig('case_4_scatter_chart.png', dpi=(300))
plt.show()
```
