## Line Charts
### Data Visualization with Python example

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Method 1

```python
#Hard-coded data for illustration purposes.
temps = {"January": {'Warsaw':-2, 'Miami':25},
         "February": {'Warsaw':0, 'Miami':26},
         "March": {'Warsaw':3, 'Miami':27},
         "April": {'Warsaw':9, 'Miami':29},
         "May": {'Warsaw':14, 'Miami':31},
         "June": {'Warsaw':18, 'Miami':32},
         "July": {'Warsaw':20, 'Miami':33},
         "August": {'Warsaw':19, 'Miami':33},
         "September": {'Warsaw':14, 'Miami':32},
         "October": {'Warsaw':9, 'Miami':30},
         "November": {'Warsaw':4, 'Miami':28 },
         "December": {'Warsaw':0, 'Miami':26}}

df_temps = pd.DataFrame.from_dict(temps, orient='index')

plt.style.use('seaborn')

df_temps.plot(kind='line',
              figsize=(8,4),
              ylim=(-5, 40),
              xticks=np.arange(0,12,1))
plt.xticks(rotation=90)           
plt.title('Monthly Average Temperatures')
plt.ylabel('Temperature')

plt.show()
```

### Method 2

```python
#Hard-coded data for illustration purposes.
warsaw = [-2,0,3,9,14,18,20,19,14,9,4,0]
miami = [25,26,27,29,31,32,33,33,32,30,28,26]
months = ['January','February','March','April','May','June',
          'July','August','September','October','November','December']

plt.style.use('seaborn')

figure, axes = plt.subplots()
figure.set_size_inches(8,4)

axes.plot(months, warsaw, label='Warsaw')
axes.plot(months, miami, label='Miami')
axes.set(ylim=(-5,40),
         xticks=np.arange(0,12,1))
axes.legend(loc='upper right')     
axes.set_xticklabels(months)

plt.xticks(rotation=90) 
plt.title('Monthly Average Temperatures')
plt.ylabel('Temperature')

plt.savefig('case_1_line_chart.png', dpi=(300), bbox_inches='tight')
plt.show()
```
