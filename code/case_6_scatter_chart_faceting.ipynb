## Scatter Charts
### Data Visualization with Python example

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load data
attributes = ['abalone', 'length', 'weight', 'rings']
abalone = pd.read_csv('abalone.data', header=None, names=attributes, usecols=[0, 1,4,8])
```

```python
m = abalone[abalone['abalone'] == 'M']
f = abalone[abalone['abalone'] == 'F']
i = abalone[abalone['abalone'] == 'I']
```

```python
#M
m_x1 = m.length
m_y1 = m.weight
m_r = m.rings

#F
f_x2 = f.length
f_y2 = f.weight
f_r = f.rings

#I
i_x3 = i.length
i_y3 = i.weight
i_r = i.rings
```

```python
#Plot data
plt.style.use('default')

figure, axes = plt.subplots(3)
figure.set_size_inches(5,10)

axes[0].set_title('Correlation of Length and Weight, focus on M', fontsize=12)
axes[0].set_xlabel('Length', fontsize=10)
axes[0].set_ylabel('Weight', fontsize=10)
axes[0].scatter(m_x1, m_y1, s=m_r, color='darkorange', label='M')
axes[0].scatter(f_x2, f_y2, s=f_r, color='dimgray', label='F')
axes[0].scatter(i_x3, i_y3, s=i_r, color='dimgray', label='I')
axes[0].legend(loc='upper left')

axes[1].set_title('Correlation of Length and Weight, focus on F', fontsize=12)
axes[1].set_xlabel('Length', fontsize=10)
axes[1].set_ylabel('Weight', fontsize=10)
axes[1].scatter(m_x1, m_y1, s=m_r, color='dimgray', label='M')
axes[1].scatter(f_x2, f_y2, s=f_r, color='darkorange', label='F')
axes[1].scatter(i_x3, i_y3, s=i_r, color='dimgray', label='I')
axes[1].legend(loc='upper left')

axes[2].set_title('Correlation of Length and Weight, focus on I', fontsize=12)
axes[2].set_xlabel('Length', fontsize=10)
axes[2].set_ylabel('Weight', fontsize=10)
axes[2].scatter(m_x1, m_y1, s=m_r, color='dimgray', label='M')
axes[2].scatter(f_x2, f_y2, s=f_r, color='dimgray', label='F')
axes[2].scatter(i_x3, i_y3, s=i_r, color='darkorange', label='I')
axes[2].legend(loc='upper left')

plt.tight_layout()
plt.savefig('case_6_scatter_chart_faceting.png', dpi=(300))
plt.show()
```
