## Box Charts
### Data Visualization with Python example

```python
import matplotlib.pyplot as plt
```

```python
#Load data
input = open('input_data_9.txt')
input_lines = input.readlines()
input.close()

#Step by step transformation for clarity
data = [el for el in input_lines if '=' in el] 
data = [x[7:] for x in data]
data = [x.split(')') for x in data]
data = [x[0] for x in data]
data = [x.split(',') for x in data]

data = [list(map(float, x)) for x in data]
```

```python
#Plot data
plt.style.use('seaborn')

figure, axes = plt.subplots(1,2)
figure.set_size_inches(12,4)

labels = ['d'+str(x) for x in range(1,9)]

#Dataset 1
axes[0].boxplot(data[:8], labels=labels)
axes[0].set_title('Dataset 1', fontsize=14)
axes[0].set(xlabel='Available strategies', ylabel='Values')
axes[0].set(ylim=(0, 0.2))

#Dataset 2
axes[1].boxplot(data[8:], labels=labels)
axes[1].set_title('Dataset 2', fontsize=14)
axes[1].set(xlabel='Available strategies', ylabel='Values')
axes[1].set(ylim=(0, 0.2))

axes[0].label_outer()
axes[1].label_outer()

plt.tight_layout()
plt.savefig('case_9_box_chart.png', dpi=(300))
plt.show()
```
