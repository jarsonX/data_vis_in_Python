```python
#Samples
sample_1 = set_1[0:30]
sample_2 = set_2[0:30]
sample_3 = set_3[0:30]

N = range(30)
```

```python
#Values for s parameter
#Values should fit into one of 4 categories
#Dataset is automatically divided into quartiles (4 labels)
def create_bins(dataset, sample=4, labels=[1,2,3,4]):
    dataset_ranges = pd.cut(dataset, sample, labels=labels)
    return [x*30 for x in dataset_ranges]

s_1 = create_bins(sample_1)
s_2 = create_bins(sample_2)
s_3 = create_bins(sample_3)
```

```python
#Plot data
figure, axes = plt.subplots(3)
figure.set_size_inches(5,10)

axes[0].set_title('All sets, focus on Set 1', fontsize=12)
axes[0].set_xlabel('Sample', fontsize=10)
axes[0].set_ylabel('Result', fontsize=10)
axes[0].scatter(N, sample_1, s=s_1, color='darkorange')
axes[0].scatter(N, sample_2, s=s_2, color='dimgray')
axes[0].scatter(N, sample_3, s=s_3, color='dimgray')

axes[1].set_title('All sets, focus on Set 2', fontsize=12)
axes[1].set_xlabel('Sample', fontsize=10)
axes[1].set_ylabel('Result', fontsize=10)
axes[1].scatter(N, sample_1, s=s_1, color='dimgray')
axes[1].scatter(N, sample_2, s=s_2, color='darkorange')
axes[1].scatter(N, sample_3, s=s_3, color='dimgray')

axes[2].set_title('All sets, focus on Set 3', fontsize=12)
axes[2].set_xlabel('Sample', fontsize=10)
axes[2].set_ylabel('Result', fontsize=10)
axes[2].scatter(N, sample_1, s=s_1, color='dimgray')
axes[2].scatter(N, sample_2, s=s_2, color='dimgray')
axes[2].scatter(N, sample_3, s=s_3, color='darkorange')

plt.tight_layout()
plt.savefig('case_11_scatter_chart.jpg', dpi=(300))
plt.show()
```
