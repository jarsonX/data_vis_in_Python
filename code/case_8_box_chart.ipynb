## Box Charts
### Data Visualization with Python example

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load data
df = pd.read_csv('input_data_8.csv', sep=';', decimal=',')
```

```python
#Plot data
box_1 = df.loc[0]
box_2 = df.loc[24]
box_3 = df.loc[49]
box_4 = df.loc[86]

all = [box_1, box_2, box_3, box_4]

plt.style.use('seaborn')

figure, axes = plt.subplots()
figure.set_size_inches(8,10)

axes.boxplot(all, labels=['1st','25th','75th','Last'])
axes.set_title('Solutions in the 1st, 25th, 50th and the last iteration', fontsize=14)
axes.set(xlabel='Iteration', ylabel='Solution')

plt.tight_layout()
plt.savefig('case_8_box_chart.png', dpi=(300))
plt.show()
```
